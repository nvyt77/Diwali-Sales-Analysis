# Diwali Sales Analysis

# Importing Libraries

```python
# import python libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

## Import Dataset

```python
# import csv file
df=pd.read_csv('Diwali Sales Data.csv',encoding='unicode-escape')
```

```python
df.info()
```

## Data Cleaning

```python
#drop unrelated/blank columns
df.drop(['Status','unnamed1'], axis=1, inplace=True)
```

```python
df.info()
```

```python
pd.isnull(df)
```

```python
#check for null values
pd.isnull(df).sum()
```

```python
df.shape
```

```python
# drop null values
df.dropna(inplace=True)
```

```python
df.shape
```

```python
pd.isnull(df).sum()
```

```python
# change data type
df['Amount']=df['Amount'].astype('int')
```

```python
df['Amount'].dtypes
```

```python
# use describe() for specific columns
df[['Age','Orders','Amount']].describe()
```

## Exploratory Data Analysis

## Gender

```python
# plotting a bar chart for Gender and it's count
Gender_count=sb.countplot(data=df,x='Gender',palette='magma')
for bars in Gender_count.containers:
    Gender_count.bar_label(bars)  
sb.set(rc={'figure.figsize':(5,5)})
sb.set_style('white')
font1={'family':'fantasy','size':25,'color':'purple'}
plt.title('Count of Gender', fontdict=font1)
```

```python
# plotting a bar chart for gender vs total amount
sales_amount=df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sb.barplot(x='Gender',y='Amount',data=sales_amount, palette='magma')
for index, value in enumerate(sales_amount['Amount']):
    plt.text(index, value, f'{value}', ha='center', va='bottom', color='black')
sb.set(rc={'figure.figsize':(5,5)})
sb.set_style('white')
font1={'family':'fantasy','size':25,'color':'purple'}
plt.title('Gender Versus Total Amount', fontdict=font1)
```

***Obervation 1: From the above 2 graphs, we can see that most of the buyers are females and even the purchasing power of females are greater than males.***

## Age

```python
# plotting a bar chart for Age group and it's count
age_group_count=sb.countplot(data=df,x='Age Group',hue='Gender', palette='tab10')
for bars in age_group_count.containers:
    age_group_count.bar_label(bars)
sb.set(rc={'figure.figsize':(15,5)})
sb.set_style('white')
font1={'family':'fantasy','size':25,'color':'purple'}
plt.title('Count of Age Group', fontdict=font1)
```

```python
# plotting a bar chart for Age group vs total amount
sales_amount2=df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sb.barplot(x='Age Group', y='Amount', data=sales_amount2, palette='tab10')
for index, value in enumerate(sales_amount2['Amount']):
    plt.text(index, value, f'{value}', ha='center', va='bottom', color='black')
sb.set(rc={'figure.figsize':(10,5)})
sb.set_style('white')
font1={'family':'fantasy','size':25,'color':'purple'}
plt.title('Age Group Versus Total Amount', fontdict=font1)
```

***Obervation 2 :From above 2 graphs, we can see that most of the buyers are of age group between 26-35 yrs and they are female.***


## State

```python
# total number of orders from top 10 states
state_orders = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)
bar_plot=sb.barplot(data=state_orders, x ='State', y='Orders', palette='Set1')
for p in bar_plot.patches:
    bar_plot.annotate(f'{int(p.get_height())}', (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='bottom', 
    fontsize=12, color='black')
sb.set(rc={'figure.figsize':(19,5)})
sb.set_style('white')
font1={'family':'fantasy','size':25,'color':'purple'}
plt.title('Top 10 States Versus Orders', fontdict=font1)
```

```python
# total amount from top 10 states
sales_amount3 = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
sb.barplot(data = sales_amount3, x = 'State',y= 'Amount', palette='Set1')
for index, value in enumerate(sales_amount3['Amount']):
    plt.text(index, value, f'{value}', ha='center', va='bottom', color='black')
sb.set(rc={'figure.figsize':(19,5)})
sb.set_style('white')
font1={'family':'fantasy','size':25,'color':'purple'}
plt.title('Total Amount Versus Top 10 States', fontdict=font1)
```

***Obervation 3 :From above 2 graphs, we can see that most of the orders and total amount are from Uttar Pradesh, Maharashtra and Karnataka respectively.***


## Marital Status

```python
# plotting a bar chart for Martial Status and it's count
MartialStatus_count=sb.countplot(data=df,x='Marital_Status',palette='Set1')
for bars in MartialStatus_count.containers:
    MartialStatus_count.bar_label(bars)  
sb.set(rc={'figure.figsize':(5,5)})
sb.set_style('white')
font1={'family':'fantasy','size':25,'color':'purple'}
plt.title('Count of Martial Status', fontdict=font1)
```

```python
# plotting a bar chart for Marital Status vs total amount
sales_amount4=df.groupby(['Marital_Status','Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
bar_plot1=sb.barplot(x='Marital_Status',y='Amount',data=sales_amount4, palette='Set1',hue='Gender')
for bar in bar_plot1.patches:
    bar_height = bar.get_height()
    bar_x = bar.get_x() + bar.get_width() / 2  # Calculate x position for centered alignment
    plt.text(bar_x, bar_height, f'{bar_height:.0f}', ha='center', va='bottom', color='black') 
sb.set(rc={'figure.figsize':(5,5)})
sb.set_style('white')
font1={'family':'fantasy','size':25,'color':'purple'}
plt.title('Marital Status Versus Total Amount', fontdict=font1)
```

***Observation 4: From above 2 graphs, we can see that most of the buyers are married (women) and they have high purchasing power.***

## Occupation

```python
# plotting a bar chart for Occupation and it's count

# Count the occurrences of each occupation and sort them
occupation_counts = df['Occupation'].value_counts()
sorted_occupations = occupation_counts.index.tolist()

# Plotting a bar chart for Occupation and its count, ordered by count
Occup_count = sb.countplot(data=df, x='Occupation', order=sorted_occupations, palette='tab10')

# Adding bar labels
for bars in Occup_count.containers:
    Occup_count.bar_label(bars)
sb.set(rc={'figure.figsize':(19,5)})
sb.set_style('white')
font1={'family':'fantasy','size':25,'color':'purple'}
plt.title('Count of Occupation', fontdict=font1)
```

```python
# plotting a bar chart for Occupation vs total amount
sales_amount5=df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sb.barplot(x='Occupation', y='Amount', data=sales_amount5, palette='tab10')
for index, value in enumerate(sales_amount5['Amount']):
    plt.text(index, value, f'{value}', ha='center', va='bottom', color='black')
sb.set(rc={'figure.figsize':(19,5)})
sb.set_style('white')
font1={'family':'fantasy','size':25,'color':'purple'}
plt.title('Occupation Versus Total Amount', fontdict=font1)
```

***Observation 5: From above 2 graphs, we can see that most of the buyers are working in IT, Healthcare and Aviation sector.***


## Product Category

```python
# plotting a bar chart for Top 10 Product Category and it's count
prod_top_categ=df['Product_Category'].value_counts().nlargest(10).index
filtered_df=df[df['Product_Category'].isin(prod_top_categ)]
prod_categ=sb.countplot(data=filtered_df, x='Product_Category',palette='Set1', order=prod_top_categ)
for bars in prod_categ.containers:
    prod_categ.bar_label(bars)
sb.set(rc={'figure.figsize':(19,5)})
sb.set_style('white')
font1={'family':'fantasy','size':25,'color':'purple'}
plt.title('Count of Top 10 Product Category', fontdict=font1)
```

```python
# plotting a bar chart for Top 10 Product Category vs total amount
sales_amount6=df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
sb.barplot(data=sales_amount6, x='Product_Category', y='Amount', palette='Set1')
for index, value in enumerate(sales_amount6['Amount']):
    plt.text(index, value, f'{value}', ha='center', va='bottom', color='black')
sb.set(rc={'figure.figsize':(19,5)})
sb.set_style('white')
font1={'family':'fantasy','size':25,'color':'purple'}
plt.title('Product Category Versus Total Amount', fontdict=font1)
```

***Observation 6: From above 2 graphs, we can see that most of the sold products are from "Food", "Clothing & Apparel" and "Electronics & Gadgets" category.***


## Conclusion


***The analysis of the provided graphs indicates that the primary demographic of buyers consists predominantly of married women aged 26-35, particularly from Uttar Pradesh, Maharashtra, and Karnataka, who possess higher purchasing power and are mostly employed in the IT, Healthcare, and Aviation sectors.***
